```python
def label_encoder(df, column_name):
    label_encoder = preprocessing.LabelEncoder()

    df[column_name]= label_encoder.fit_transform(df[column_name])
    
    return df[column_name], label_encoder
```

```python
df['POSTED_BY'], label_encoder_posted_by = label_encoder(df, 'POSTED_BY')
df['BHK_OR_RK'], label_encoder_posted_by = label_encoder(df, 'BHK_OR_RK')
df['ADDRESS'],label_encoder_posted_by = label_encoder(df, 'ADDRESS')
```

```python
#Features selection
from sklearn.feature_selection import VarianceThreshold
x = VarianceThreshold(threshold=0.01)
x.fit(df)
```

```python
drop_col = [column for column in df.columns if column not in df.columns[x.get_support()]]
drop_col
```

```python
df_col= df.drop(drop_col, axis = 1)
df_col
```

```python
X = df_col[['POSTED_BY', 'UNDER_CONSTRUCTION', 'RERA', 'BHK_NO.',
       'SQUARE_FT', 'READY_TO_MOVE', 'RESALE', 'ADDRESS', 'LONGITUDE',
       'LATITUDE']]
Y = df_col[['TARGET(PRICE_IN_LACS)']]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
ltr = lr.fit(X,Y)
print(ltr.score(X, Y))
```

```python
mselr= LinearRegression()
mselr.fit(X,Y)
from sklearn.metrics import mean_squared_error as MSE
rmse = (np.sqrt(MSE(Y, mselr.predict(X))))
print("Root Mean Squared Error: ",rmse)
```

```python
from sklearn.tree import DecisionTreeRegressor

dtr = DecisionTreeRegressor().fit(X,Y)
print(dtr.score(X, Y))
```

```python
Dtr = DecisionTreeRegressor()
Dtr.fit(X_train,Y_train)
dmse = (np.sqrt(MSE(Y_test, mselr.predict(X_test))))
print("Root Mean Squared Error: ",dmse)
```
